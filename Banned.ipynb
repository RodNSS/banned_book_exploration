```python
import pandas as pd
import requests

url = "https://www.googleapis.com/books/v1/volumes"

book_data = []

for title in dfs[0]['publication']:
    params = {
        "q": f"title:{title}",
        "projection": "full"
    }
    response = requests.get(url, params=params)
    data = response.json()
    
    if 'items' not in data:
        publication_date = None
        topic_description = None
    else:
        publication_date = data['items'][0]['volumeInfo'].get('publishedDate')
        topic_description = data['items'][0]['volumeInfo'].get('description')
    
    book_info = {
        'publication': title,
        'publication_date': publication_date,
        'topic_description': topic_description
    }
    book_data.append(book_info)

df_book_data = pd.DataFrame(book_data)
dfs[0] = pd.concat([dfs[0], df_book_data[['publication_date', 'topic_description']]], axis=1)
dfs[0]
```
